```python
import numpy as np  # array manipulation
import matplotlib.pyplot as plt  # plotting
```

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size=5>
		ورودی: 
	</font>
</div>

```python
x = np.array(list(map(float, input().split()))) 
```

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size=5>
		تنظیم پارامترها: 
	</font>
</div>

```python
population = 1000

generations = 150

mu = 0.1 # mutation probapility

chromosomes = 20*np.random.rand(population, x.shape[0])-10 # uniform random between -10 and 10
```

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size=5>
		یک الگوریتم ژنتیک شامل سه بخش است:
        <ol>
            <li>
             selection
            </li>
            <li>
             crossover
            </li>
            <li>
             mutation
            </li>
        </ol>
        در اول بخش که 
        selection 
        می‌باشد ابتدا باید میزان 
        fitness
        هر کروموزوم محاسبه شود و سپس بر اساس آن انتخاب صورت بگیرد. مقدار 
        fitness
        را برابر با 
        $y$
        در نظر می‌گیریم. از آنجایی که احتمال انتخاب هر کروموزوم باید با
        fitness 
        آن متناسب باشد، آن را به این شکل محاسبه می‌کنیم:
        $$
        p_i = \frac{\mathrm{fitness}_i-\min(\mathrm{fitness})}{\sum_i\left(\mathrm{fitness}_i-\min(\mathrm{fitness})\right)}
        $$
	</font>
</div>

```python
def fitness(w):
    return np.dot(w, x)


def selection(fitness): # returns n indices that are chosen for crossover
    probs = (fitness - np.min(fitness))/sum(fitness - np.min(fitness))
    return np.random.choice(population, population, p=probs)
```

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size=5>
		در 
        crossover
        نیز دو کروموزوم از یک
        pivot 
        تصادفی با هم جا به جا می‌شوند. در 
        mutation
        یکی از عناصر کروموزوم به طور تصافی تغییر می‌کند.
	</font>
</div>

```python
def crossover(w1, w2):
    pivot = np.random.randint(0, w1.shape[0])
    new1 = w1
    new2 = w2
    new1[:pivot+1] = w2[:pivot+1]
    new2[:pivot+1] = w1[:pivot+1]
    return new1, new2


def mutate(w):
    index = np.random.randint(0, w.shape[0])
    w[index] = np.random.rand()*20 - 10
```

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size=5>
		در نهایت این توابع را در کنار هم به کار می‌بریم.
	</font>
</div>

```python
best_fitness = []

for i in range(generations):
    chosen = chromosomes[selection(fitness(chromosomes))]  # selection
    new_gen = np.empty((population, x.shape[0]))
    
    for j in range(0, population, 2):  # crossover
        new_gen[j], new_gen[j+1] = crossover(chromosomes[j], chromosomes[j+1])
        
    for j in range(population):  # mutation
        if np.random.rand() < mu:
            mutate(new_gen[j])
            
    new_f = fitness(new_gen)
            
    print(f"generation {i+1} best fitness: {np.max(new_f)}\ncorresponding weights: {new_gen[np.argmax(new_f)]}\n")
    best_fitness.append(np.max(new_f))
    chromosomes = new_gen

plt.plot(best_fitness)
plt.xlabel("generation")
plt.ylabel("best fit")
plt.show()
    
```
